```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
O = []
M = []

b,e = 2,10000

r = range(b,int(e),1)

for i in r:
    o = 0
    m = 0

    if i**0.5 == int(i**0.5): 
        o=1 
        m=1
    for k in range(2,int((i**0.5))):
        o += 1
        if i%k == 0:
            o += 2
            m += 2
        
    O.append(o)
    M.append(m)
```

```python
def tr(O, treshold):

    Ox, Oy = [0],[O[0]]
    for j in range(1, len(O)):
        if (O[j]>=O[Ox[-1]]) and (O[j]-Oy[-1] <= treshold ) and (j > Ox[-1]):
            Ox.append(j)
            Oy.append(O[j])
        
    return Ox, Oy
```

```python
tr([1,2,3,6,10,1,6,1,7,1,8,4,5,6], 3)
```

```python
Ox, Oy = tr(O, 10000)
Mx, My = tr(M, 10000)
```

```python
# Ox, Oy = range(2,10000), O
# Mx, My = Ox, M
```

```python
fig, ax = plt.subplots(figsize=(9,9))
# ax.plot(rO,O,label='Time')
lspc = np.linspace(b,e)
ax.plot(Ox,Oy,label = 'Time_Avg')
ax.plot(lspc, np.sqrt(3*lspc), label = 'sqrt(3n)')
ax.plot(lspc, np.sqrt(lspc), label = 'sqrt(n)')
ax.plot(Mx,My,label = 'Mem_Avg')
ax.legend()
```

```python
O = []
M = []

b,e = 2,10000

r = range(b,int(e),1)

for i in r:
    j = i

    o = 0
    m = 0

    d = 2
    
    while j>2:
        if j%d!=0: 
            m+=1
            d+=1
            continue
        while j%d==0:
            j//=d
            o += 1
            m += 1
        if d>i**0.5+1:
            break
        d+=1

        
        
    O.append(o)
    M.append(m)
```

```python
Ox, Oy = tr(O, 10000)
Mx, My = tr(M, 10000)
```

```python
fig, ax = plt.subplots(figsize=(9,9))
# ax.plot(rO,O,label='Time')
lspc = np.linspace(b,e)
ax.plot(Ox,Oy,label = 'Time_Avg')
ax.plot(lspc, 2*np.log(lspc), label = 'sqrt(n)')
ax.plot(lspc, np.log(lspc), label = 'sqrt(n)')
ax.plot(Mx,My,label = 'Mem_Avg')
ax.legend()
```
